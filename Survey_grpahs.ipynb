```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Language frequencies data
languages = [
    'Spanish (es)', 'Chinese (zh)', 'Arabic (ar)', 'French (fr)', 'German (de)',
    'Italian (it)', 'Portuguese (pt)', 'Russian (ru)', 'Korean (ko)',
    'Dutch (nl)', 'Polish (pl)', 'Swedish (sv)', 'Ukrainian (uk)', 'Persian (fa)',
    'Hindi (hi)', 'Bengali (bn)', 'Telugu (te)', 'Japanese (ja)'
]
frequencies = [8, 7, 6, 5, 4, 4, 4, 4, 3, 3, 3, 3, 3, 3, 2, 1, 1, 1]

# Define the categories based on previous categorization
latin_based = ['Spanish (es)', 'French (fr)', 'German (de)', 'Italian (it)', 'Portuguese (pt)', 'Russian (ru)', 'Dutch (nl)', 'Polish (pl)', 'Swedish (sv)', 'Ukrainian (uk)']
cjk = ['Chinese (zh)', 'Japanese (ja)', 'Korean (ko)']
other_non_latin = ['Arabic (ar)', 'Persian (fa)', 'Bengali (bn)', 'Hindi (hi)', 'Telugu (te)']

# Define colors and hatches for each category
colors = []
hatches = []

for lang in languages:
    if lang in latin_based:
        colors.append('skyblue')  # Color for Latin-based languages
        hatches.append('')  # No hatch for Latin-based
    elif lang in cjk:
        colors.append('orange')  # Color for CJK languages
        hatches.append('///')  # Diagonal hatch for CJK
    elif lang in other_non_latin:
        colors.append('green')  # Color for other non-Latin scripts
        hatches.append('xx')  # Cross hatch for other non-Latin scripts

# Create the bar chart with different colors and hatches
plt.figure(figsize=(7, 4))  # Adjust figure size for double-column papers
bars = plt.bar(languages, frequencies, color=colors)

# Apply hatch patterns
for bar, hatch in zip(bars, hatches):
    bar.set_hatch(hatch)

# Tilt x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set labels
plt.ylabel('Number of Approaches')

# Create legend handles, ensuring both color and hatch patterns are represented
latin_patch = mpatches.Patch(facecolor='skyblue', edgecolor='black', label='Latin-based')
cjk_patch = mpatches.Patch(facecolor='orange', edgecolor='black', hatch='///', label='CJK')
non_latin_patch = mpatches.Patch(facecolor='green', edgecolor='black', hatch='xx', label='Other non-Latin')

# Add legend to the top-right corner
plt.legend(handles=[latin_patch, cjk_patch, non_latin_patch], loc='upper right')

# Save the figure as a .pdf file in high resolution suitable for research papers
plt.tight_layout()
plt.savefig("language_frequency_chart.pdf", format="pdf", bbox_inches='tight')  # Save as PDF

# Show the plot (optional)
plt.show()


# Show the pl
```
